```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.tri as tri
config = {
"font.family": "sans-serif",
"font.size": 18,
"font.sans-serif": ["Microsoft YaHei"],
"xtick.direction": "out",
"ytick.direction": "out",
"axes.unicode_minus": False,
"figure.facecolor" : "none",
"axes.facecolor" : "none",
"svg.fonttype":"none",
"svg.hashsalt":"hell0"}
plt.rcParams.update(config)
```

```python
path = './egs_1/data/'
filenames = os.listdir(path)

index = 4
d = np.loadtxt(path+filenames[index])
print(path+filenames[index])
#x, y, p, r_t, f_t
x = d[:-1,0]
y = d[:-1,1]
p = d[:-1,2]/1e6
r_t = d[:-1,3]
f_t = d[:-1,4]

triangulation = tri.Triangulation(x, y)

fig,axs = plt.subplots(1,3,figsize=(15,4))

# 英文标注
# cs1 = axs[0].tricontourf(triangulation, p, cmap='coolwarm',levels = np.linspace(6,10,21))
# cb1 = fig.colorbar(cs1,format = '%.0f',label='Pressure (MPa)',ticks=[6,7,8,9,10]) 

# cs2 = axs[1].tricontourf(triangulation, r_t, cmap='coolwarm',levels = np.linspace(398,500,21))
# cb2 = fig.colorbar(cs2,format = '%.0f',label='Rock $T$ (K)',ticks=[400,425,450,475,500]) 

# cs3 = axs[2].tricontourf(triangulation, f_t, cmap='coolwarm',levels = np.linspace(300,500,21))
# cb3 = fig.colorbar(cs3,format = '%.0f',label='Fluid $T$ (K)',ticks=[300,350,400,450,500]) 

# 中文标注
cs1 = axs[0].tricontourf(triangulation, p, cmap='coolwarm',levels = np.linspace(6,10,21))
cb1 = fig.colorbar(cs1,format = '%.0f',label='压力（MPa）',ticks=[6,7,8,9,10]) 

cs2 = axs[1].tricontourf(triangulation, r_t, cmap='coolwarm',levels = np.linspace(398,500,21))
cb2 = fig.colorbar(cs2,format = '%.0f',label='岩石温度（K）',ticks=[400,425,450,475,500]) 

cs3 = axs[2].tricontourf(triangulation, f_t, cmap='coolwarm',levels = np.linspace(300,500,21))
cb3 = fig.colorbar(cs3,format = '%.0f',label='流体温度（K）',ticks=[300,350,400,450,500]) 

for ax in axs:
    ax.set_xlim(0,100)
    ax.set_ylim(0,100)
    ax.set_xticks([0,50,100])
    ax.set_yticks([0,50,100])
    # ax.spines['right'].set_visible(False)
    # ax.spines['top'].set_visible(False)
    # ax.spines['left'].set_visible(False)
    # ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_color('gray')
    ax.spines['top'].set_color('gray')
    ax.spines['left'].set_color('gray')
    ax.spines['bottom'].set_color('gray')
    ax.set_xlabel('x (m)')
    ax.tick_params(axis='x', color='w',length=2)
    ax.tick_params(axis='y', color='w',length=2)
    3
axs[0].set_ylabel('y (m)')

plt.tight_layout()

```
